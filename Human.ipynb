# **XGBoost**

```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import StandardScaler
import numpy as np
import time

# Load data
data = pd.read_csv('/content/drive/MyDrive/human_df.csv', header=0)
columns_to_drop = ['Ligand SMILES code']
data.drop(columns_to_drop, axis=1, inplace=True)

# Separate features and target
y = data['Affinity (nm) IC50']  # Assuming the target is 'Affinity (nm) IC50'
X = data.drop(['Affinity (nm) IC50'], axis=1)

# Initialize XGBRegressor
base_model = xgb.XGBRegressor(
    n_jobs=4,
    objective='reg:squarederror',  # Use 'reg:squarederror' for regression
    use_label_encoder=False
)

# Initialize Leave-One-Out cross-validator
loo = LeaveOneOut()

# Store cross-validation results
mse_scores = []
fitting_times = []
inference_times = []

# Perform LOO-CV
for train_index, test_index in loo.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize the scaler
    scaler = StandardScaler()

    # Fit the scaler on the training data and transform both training and test data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the model on the scaled training data
    start = time.time()
    modelfit = base_model.fit(X_train_scaled, y_train, eval_metric='rmse')
    fittingTime = time.time() - start
    fitting_times.append(fittingTime)

    # Make predictions
    start = time.time()
    predictions = modelfit.predict(X_test_scaled)
    inferenceTime = time.time() - start
    inference_times.append(inferenceTime)

    # Calculate metrics
    mse = mean_squared_error(y_test, predictions)
    mse_scores.append(mse)

# Calculate average metrics across all folds
average_mse = np.mean(mse_scores)
average_fitting_time = np.mean(fitting_times)
average_inference_time = np.mean(inference_times)

print('Average Mean Squared Error (MSE) = {:.4f}'.format(average_mse))
print('Average Fitting Time = {:.4f} seconds'.format(average_fitting_time))
print('Average Inference Time = {:.4f} seconds'.format(average_inference_time))
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression, RFE
import numpy as np
import time

# Load data
data = pd.read_csv('/content/drive/MyDrive/human_df.csv', header=0)
columns_to_drop = ['Ligand SMILES code']
data.drop(columns_to_drop, axis=1, inplace=True)

# Separate features and target
y = data['Affinity (nm) IC50']  # Assuming the target is 'Affinity (nm) IC50'
X = data.drop(['Affinity (nm) IC50'], axis=1)

# Step 1: Initial feature selection using SelectKBest
initial_feature_count = 100  # Adjust this based on your specific needs
select_k_best = SelectKBest(score_func=f_regression, k=initial_feature_count)
X_new = select_k_best.fit_transform(X, y)

# Initialize XGBRegressor
base_model = xgb.XGBRegressor(
    n_jobs=4,
    objective='reg:squarederror',  # Use 'reg:squarederror' for regression
    eval_metric='rmse',  # Set eval_metric in constructor
    use_label_encoder=False
)

# Initialize Leave-One-Out cross-validator
loo = LeaveOneOut()

# Store cross-validation results
mse_scores = []
mae_scores = []
fitting_times = []
inference_times = []

# Perform LOO-CV
for train_index, test_index in loo.split(X_new):
    X_train, X_test = X_new[train_index], X_new[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize the scaler
    scaler = StandardScaler()

    # Fit the scaler on the training data and transform both training and test data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Step 2: RFE
    n_features_to_select = 10  # Adjust this based on your specific needs
    rfe = RFE(estimator=base_model, n_features_to_select=n_features_to_select)
    X_train_rfe = rfe.fit_transform(X_train_scaled, y_train)
    X_test_rfe = rfe.transform(X_test_scaled)

    # Train the model on the selected features
    start = time.time()
    modelfit = base_model.fit(X_train_rfe, y_train)
    fittingTime = time.time() - start
    fitting_times.append(fittingTime)

    # Make predictions
    start = time.time()
    predictions = modelfit.predict(X_test_rfe)
    inferenceTime = time.time() - start
    inference_times.append(inferenceTime)

    # Calculate metrics
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    mse_scores.append(mse)
    mae_scores.append(mae)

# Calculate average metrics across all folds
average_mse = np.mean(mse_scores)
average_mae = np.mean(mae_scores)
average_fitting_time = np.mean(fitting_times)
average_inference_time = np.mean(inference_times)

print('Average Mean Squared Error (MSE) = {:.4f}'.format(average_mse))
print('Average Mean Absolute Error (MAE) = {:.4f}'.format(average_mae))
print('Average Fitting Time = {:.4f} seconds'.format(average_fitting_time))
print('Average Inference Time = {:.4f} seconds'.format(average_inference_time))
```
